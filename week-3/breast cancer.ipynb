# heat map of csv file

```python

print("Columns in df2:", df2.columns)


diagnosis_cols = [col for col in df2.columns if 'diagnosis' in col.lower()]
print("Columns containing 'diagnosis':", diagnosis_cols)


print(df2.head())


if 'diagnosis_M' in df2.columns:
    df2['diagnosis_M'].value_counts().plot()
elif 'diagnosis_m' in df2.columns:
    df2['diagnosis_m'].value_counts().plot()
elif len(diagnosis_cols) > 0:
    
    df2[diagnosis_cols[0]].value_counts().plot()
else:
    print("No 'diagnosis_M' or similar column found.")

plt.show()

```
